```python
import requests as re
import datetime
import time
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import pickle
```

```python
api_key = "YOUR_API_KEY_HERE"
commute_data = pd.read_csv('../data/commute_data.csv')
api_endpoint = 'https://maps.googleapis.com/maps/api/distancematrix/json'
```

```python
results = []
for ind,row in commute_data.iterrows():
    departure_time = datetime.datetime(row.year,row.month,row.day,row.hour,row.minute)
    unix_departure_time = time.mktime(departure_time.timetuple())
    print(unix_departure_time)
    print(departure_time)
    
    formatted_departure_time = departure_time.strftime('%Y-%m-%dT%H:%M:%S')
    origin='{lat},{lng}'.format(lat=row.orig_lat,lng=row.orig_lng)
    destination = '{lat},{lng}'.format(lat=row.dest_lat,lng=row.dest_lng)
    request_url = f"{api_endpoint}?origins={origin}&destinations={destination}&departure_time={int(unix_departure_time)}&key={api_key}"
    result = re.get(request_url).json()
    commute_data.loc[ind,'duration_in_traffic_seconds'] = result['rows'][0]['elements'][0]['duration_in_traffic']['value']
    
    results.append(result)
```

```python
def func_get_datetime(row):
    return datetime.datetime(row.year,row.month,row.day,row.hour,row.minute)

commute_data['duration_in_traffic_minutes'] = commute_data['duration_in_traffic_seconds']/60
commute_data['datetime'] = commute_data.apply(func_get_datetime,1)
commute_data['time'] = commute_data.datetime.apply(lambda x: str(x.time()))

```

```python
commute_data.to_pickle('../data/commute_data.pickle')
with open('../data/results_list.pickle','wb') as f:
    pickle.dump(results,f)
```

```python
commute_data_morning = commute_data[commute_data['journey']=='morning']
commute_data_evening = commute_data[commute_data['journey']=='evening']
```

```python
#Morning commute - lineplot 
plt.clf()
plt.figure(figsize=(22, 10))
plt.title('Time in traffic vs when you leave home',fontdict={'fontsize':30})
sns.set_style('darkgrid',{"axes.facecolor": ".9"})
sns.set_context("notebook",font_scale=2,rc={'lines.linewidth':2.2})
sns.lineplot(data=commute_data_morning,x='time',y='duration_in_traffic_minutes',hue='week_day',palette='deep')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1));
plt.tick_params(axis='x',rotation=70)
plt.xlabel('Time you leave home')
plt.ylabel('Time in traffic (minutes)')
plt.savefig('../viz/time_in_traffic_home_lineplot.png',bbox_inches='tight')
```

```python
#Evening commute - lineplot 
plt.clf()
plt.figure(figsize=(22, 10))
plt.title('Time in traffic vs when you leave work',fontdict={'fontsize':30})
sns.set_style('darkgrid',{"axes.facecolor": ".9"})
sns.set_style(rc = {'axes.bgcolour': '#c79f52'})
sns.set_context("notebook",font_scale=2,rc={'lines.linewidth':2.2})
sns.lineplot(data=commute_data_evening,x='time',y='duration_in_traffic_minutes',hue='week_day',palette='deep')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1));
plt.tick_params(axis='x',rotation=70)
plt.xlabel('Time you leave work')
plt.ylabel('Time in traffic (minutes)')
plt.savefig('../viz/time_in_traffic_work_lineplot.png',bbox_inches='tight')
# ax.set_f
```

```python

#Morning commute - heatmap

plt.clf()
plt.figure(figsize=(18, 9))
# plt.style.use('ggplot')
sns.set_style('darkgrid',{"axes.facecolor": ".9"})
sns.set_context("notebook",font_scale=1.5,rc={'lines.linewidth':2.2})

transit_morning_heatmap = commute_data_morning[['week_day','time','duration_in_traffic_minutes']]
transit_morning_heatmap = transit_morning_heatmap.pivot(columns='week_day',index='time',values='duration_in_traffic_minutes')
transit_morning_heatmap = transit_morning_heatmap[['Monday','Tuesday','Wednesday',
                        'Thursday','Friday']]

sns.heatmap(transit_morning_heatmap,
            cmap='Spectral_r',
            linewidths=0.001)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.tick_params(axis='x',rotation=0)
plt.xlabel(None)
plt.ylabel('Time you leave home')
plt.title('Time you spend in traffic & when you leave home');
plt.savefig('../viz/time_in_traffic_home_heatmap.png',bbox_inches='tight')


```
